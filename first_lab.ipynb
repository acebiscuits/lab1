```python
import random

def downloadpictures(url1, url2, folder, fullpath):
    count = 0
    for page in range(0, 37):
        url = url1 + str(page) + url2 + folder
        print(url)
        r = requests.get(url)
        soup = BeautifulSoup(r.text, 'lxml')
        blocks = soup.findAll('div', class_='serp-item__preview')
        for block in blocks:
            if (block):
                v1 = str(count)
                zero = 4 - len(v1)
                number = '0' * zero
                name = str(number) + str(count)
                preview_block = block.find('a', class_='serp-item__link').get('href')
                link = (preview_block.split('=')[3])
                link2 = link.replace('%3A', ':')
                link3 = link2.replace('%2F', '/')
                link4 = (link3.split('&')[0])
                print(link4)
                try:
                    link5 = requests.get(link4, timeout = 10)
                except:
                    continue
                if link5.status_code == 200:
                    image_byt = link5.content #ЗДЕСЬ наверно ЗАМЕНИТЬ requests.get(f'{link4}') на линк 5
                    time.sleep(random.random())
                    with open(fullpath + '\\' + folder + '\\' + name + '.jpg', 'wb') as file:
                        file.write(image_byt)
                    count += 1      

                    
```

```python
class Iterator:
    def __init__(self, c_name):
        self.c_name = c_name
        self.counter = 0
        print(c_name)

    def __next__(self):
        self_path = 'C:\\Users\\TUFman\\Desktop\\python\\python\\dataset\\' + self.c_name + '\\' + str(self.counter).zfill(4) + '.jpg'
        if(os.path.exists(self_path)):
            #print(self_path)
            self.counter += 1
            return self_path
        else:
            raise StopIteration
```

```python
import csv
iterator = Iterator('tiger')
i = 20
with open("lab_2.csv", mode="w", encoding='utf-8') as w_file:
    file_writer = csv.writer(w_file, delimiter = ",", lineterminator="\r")
    file_writer.writerow(('Абсолютный путь', 'Относительный путь', 'Класс'))
    while(i):
        self_path = next(iterator)
        file_writer.writerow([self_path])
        rel_path = 'python' + self_path.split('python')[2]
        c_name = rel_path.split('\\')[2]
        file_writer.writerow((self_path, rel_path, c_name))
        print(self_path)
        print(rel_path)
        print(c_name)
        i -= 1
```
